```python
· Modificar entrenamiento.py para permitir al jugador realizar varios intentos de entrenamiento.

Para esto lo que tenemos que hacer es crear el apartado  de entrenamientos, y en donde podamos entrenar las estadísticas (vida, fuerza…) dependerá de su estamina y subirá aleatoriamente, siendo más probable que sea un entrenamiento poco eficiente si el animal tiene poca estamina e imposible de hacer si tiene demasiada poca.

· Usar un bucle for o while que repita el entrenamiento hasta que la energía del jugador sea menor a un valor mínimo.

Aquí lo que hay que hacer es simplemente programar para que no podamos entrenar más si nuestra estamina es menor a lo requerido para poder hacerlo

· Agregar condicionales para determinar si el entrenamiento fue exitoso según la puntuación.

Aquí lo que hay que hacer vemos que es basicámente pues poner que cuanto menos estamina tenga el entreno de menor puntuación ( o al menos, probabilidad menor a ser un entrenamiento malo)

· Implementar una opción que permita salir del bucle escribiendo “fin”.

Como bien dice, crear un comando para que el jugador pueda dejar de entrenar sin necesidad a que se le gaste toda la estamina

· Añadir control de excepciones si el usuario introduce valores no numéricos.

Evitar el poder escribir con número a la hora de entrenar y que salga un mensaje de error para indicarlo
```

```python
print("Entrenamiento de mascotas")

rareza = input("Pon la rareza de tu mascota (comun, raro, epico, mitico, legendario): ")
estamina = int(input("Pon la estamina actual de tu mascota: "))

# limites según rareza
if rareza == "comun":
    limite = 30
elif rareza == "raro":
    limite = 35
elif rareza == "epico":
    limite = 40
elif rareza == "mitico":
    limite = 45
elif rareza == "legendario":
    limite = 50
else:
    print("Rareza no valida.")
    exit()

print("Escribe 'fin' para salir del entrenamiento.\n")

while estamina > 5:   # se puede entrenar mientras quede estamina
    accion = input("Pulsa ENTER para entrenar o escribe fin: ")

    if accion == "fin":
        print("Has salido del entrenamiento.")
        break

    # restamos estamina por entrenar
    estamina -= 5
    print("Has entrenado. Estamina ahora:", estamina)

    # comprobamos el rendimiento
    if estamina < limite:
        print("Rinde peor en batalla por baja estamina.")
    else:
        print("Rinde bien en batalla.")

    print("----------------------")

print("Entrenamiento terminado. No queda estamina.")
```
